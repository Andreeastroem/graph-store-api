```python
import random
import uuid

def generate_isni(name):
    # Generate unique isni based on name set to 16 digits
    random.seed(hash(name))
    isni = ''.join([str(random.randint(0, 9)) for _ in range(16)])
    return isni

def generate_sku():
    return "P" + ''.join([str(random.randint(0, 9)) for _ in range(8)])

def generate_price():
    return round(random.uniform(10, 30), 2)

def generate_isbn():
    return ''.join([str(random.randint(0, 9)) for _ in range(13)])

def generate_dimensions():
    return {
        "height": round(random.uniform(7, 10), 1),
        "width": round(random.uniform(4, 7), 1),
        "depth": round(random.uniform(0.5, 2), 1),
        "unit": "in"
    }

def generate_weight():
    return {
        "amount": round(random.uniform(0.5, 2), 1),
        "unit": "lb"
    }

def generate_publish_info():
    return {
        "publisher": {
            "name": "Fictional Publisher"
        },
        "published": {
            "date": "2023-01-01"
        }
    }
```

```python
def convert_work_to_books(work):
    books = []
    work_id = str(uuid.uuid4())
    for language in [work["languages"]["original"]] + work["languages"]["translated"]:
        for format in work["formats"]:
            book = {
                "type": "book",
                "title": language["title"],
                "sku": generate_sku(),
                "price": {
                    "amount": generate_price(),
                    "currency": "USD"
                },
                "people": {
                    "authors": [{"name": author, "isni": generate_isni(author)} for author in work["people"]["authors"]],
                    "translators": [{"name": "", "isni": ""}],
                    "illustrators": [{"name": illustrator, "isni": generate_isni(illustrator)} for illustrator in work["people"]["illustrators"]],
                    "narrators": [{"name": narrator, "isni": generate_isni(narrator)} for narrator in work["people"]["voiceActors"]],
                },
                "language": {
                    "code": language["name"][:2].lower(),
                    "name": language["name"],
                    "original": language == work["languages"]["original"]
                },
                "publishInfo": generate_publish_info(),
                "pages": random.randint(200, 500),
                "dimensions": generate_dimensions(),
                "weight": generate_weight(),
                "isbn": generate_isbn(),
                "format": format,
                "series": {
                    "name": work["series"][0] if work["series"] else ""
                },
                "classifications": {
                    "primary": work["classifications"]["primary"],
                    "secondary": work["classifications"]["secondary"]
                },
                "universes": work["universes"],
                "work": {
                    "name": work["languages"]["original"]["title"],
                    "id": work_id
                }
            }
            books.append(book)
    return books

# Convert works to books
books = []
for work in works:
    books.extend(convert_work_to_books(work))
```
